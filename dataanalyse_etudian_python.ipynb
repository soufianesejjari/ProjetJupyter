# Soufiane Sejjari - Ajebli Ahmed - Kassaoui Manal

```python
#importation de dataset avec readcsv

import csv
import matplotlib.pyplot as plt

reader = csv.DictReader(open('C:/Users/pc/Downloads/Compressed/archive/exams.csv'))
```

```python
#creation d'un dataframe / dictionnaire a partir du dataset


result = {}
for row in reader:
    for column, value in row.items():  # consider .iteritems() for Python 2
        result.setdefault(column, []).append(value)
print(result)
```

```python
keysList = list(result.keys())
print(keysList)
```

# statistics

```python
#les statistiques de notre dataframe


keysList = list(result.keys())
for l in keysList:
    re = {i:result[l].count(i) for i in result[l] }
    if (len(re)<30):

        print(l,"->",len(re), re)

    else:
        print(l,"->",len(re), )
```

# nombres des variabes par attribut

```python
#affichage des plots pour chaque attribut

for l in keysList:
    re = {i:result[l].count(i) for i in result[l] }
    if (len(re)<30):

        print("plot de",l)
        width = 0.7
        c = ['red', 'yellow', 'black', 'blue', 'orange','pink']
        plt.bar(re.keys(), re.values(), width, color=c)
        plt.show()
    else:
        print(l," ",len(re), )
        
        
```
